## Cancer data set

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

```

```python
dataset = pd.read_csv("cancer.csv")
```

```python
colors = ['r' if label == 1 else 'b' for label in dataset["Z"]]
plt.scatter(dataset["Age"], dataset["Tumor"], c=colors)
plt.xlabel("Age")
plt.ylabel("Tumor size (in cm)")
plt.show()
```
